# Handling missing values with Sk-learn

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/loan_data_set.csv')
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include="float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='mean')
si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
# Enoding.
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset['Gender'].mode()[0],inplace=True)
```

```python
dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True)
```

```python
en_data = dataset[["Gender","Married"]]
```

```python
en_data
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first')
ohe.fit_transform(en_data)
```

```python
# here we get the data into sparse matrices now convert it into array.
ohe.fit_transform(en_data).toarray()
```

```python
# now convert this into a dataframe
ar = ohe.fit_transform(en_data).toarray()
```

```python
# now convert this into a dataframe
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes"])
```
