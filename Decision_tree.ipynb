```python
import graphviz.backend as be
```

```python
from sklearn.datasets import *
from dtreeviz.trees import *
from IPython.display import Image,display_svg,SVG
```

```python
clas = tree.DecisionTreeClassifier(max_depth=2)
iris=load_iris()
```

```python
X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas)
```

# 1.Classification

```python
from dtreeviz import dtreeviz as dtreeviz_function

viz = dtreeviz_function(clas,
               X_train,
               y_train,
               target_name=iris.target_names[0], # Using the first target name as a placeholder
               feature_names=iris.feature_names,
               class_names=list(iris.target_names), # Use class_names instead of class_name
               scale = 1.5)
viz
```

# 2.Regression

```python
from sklearn.datasets import load_diabetes
from sklearn import tree
from dtreeviz import model

regr = tree.DecisionTreeRegressor(max_depth=2)
diabetes = load_diabetes()

X_train = diabetes.data
y_train = diabetes.target
regr.fit(X_train,y_train)

viz_model = model(regr,
                  X_train,
                  y_train,
                  target_name='diabetes progression',
                  feature_names=diabetes.feature_names)

viz = viz_model.view()
viz
```
