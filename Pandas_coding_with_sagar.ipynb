```python
import pandas as pd
```

```python
# How to create a dataframe.
data = {"name":['ram','shyam','ghanshyam'],
        "Age":[10,20,30],
        "city":['pune','nagpur','baramati']
        }
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
# save above file in csv formate.
df.to_csv("output.csv",index=False)
```

```python
# save above file in excel formate.
df.to_excel("output.xlsx",index=False)
```

```python
# save above file in JSON formate.
df.to_json("output.json",index = False)
```
