# *Gradient Descent*

* Gradient Descent is an optimization algorithm used to find the minimum value of a function.

* gradient means Derivative.

```python
# apply linear regression using Gradient Descent.
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4,n_features=1,n_informative=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# lest apply Gradient Descent assuming slope is constant that is m = 78.35
# and lets assume the starting value for intercept b =0.
y_pred = ((78.35 * x)+ 0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b=0')
plt.legend()
plt.show()
```

```python
m= 78.35
b = 0
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
# lets take learning rate = 0.1
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='#00a65a',label='b={}'.format(b))
plt.plot(x,y_pred,color='#A3E4D7',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteratiom 2
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='#00a65a',label='b={}'.format(b))
plt.plot(x,y_pred1,color='#A3E4D7',label='b=0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * x )+ b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label="OLS")
plt.plot(x,y_pred2,color='#A3E4D7',label='b={}'.format(b))
plt.plot(x,y_pred1,color='#A3E4D7',label='b={}'.format(b))
plt.plot(x,y_pred,color="#A3E4D7",label='b=0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.1

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y-m*x.ravel()- b)
  b = b - (lr * loss_slope)

  y_pred = m*x + b

  plt.plot(x,y_pred)
plt.scatter(x,y)
```

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x ,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
m = 20.96
```

```python
class GDRegressor:
  def __init__(self,learning_rate,epochs):
    self.m=20.96
    self.b=-120
    self.lr=learning_rate
    self.epochs = epochs

  def fit(self, x, y):
    # calculate the b using GD
    for i in range(self.epochs):
      loss_slope = -2 * np.sum(y-self.m*x.ravel()-self.b)
      self.b = self.b - (self.lr * loss_slope)
      print(loss_slope,self.b)
    print(self.b)
```

```python
gd = GDRegressor(0.01,100)
```

```python
gd.fit(x,y)
```
